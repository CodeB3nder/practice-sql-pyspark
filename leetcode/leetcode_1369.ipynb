```python
from pyspark.sql.functions import to_date

data = [
    ("Alice", "Travel", "2020-02-12", "2020-02-20"),
    ("Alice", "Dancing", "2020-02-21", "2020-02-23"),
    ("Alice", "Travel", "2020-02-24", "2020-02-28"),
    ("Bob", "Travel", "2020-02-11", "2020-02-18")
]

columns = ["username", "activity", "startDate", "endDate"]

df = spark.createDataFrame(data, columns)
df = df.withColumn("startDate", to_date("startDate")) \
       .withColumn("endDate", to_date("endDate"))

display(df)
```

```python
df.createOrReplaceTempView("UserActivity")
```

```python
spark.sql("""
        select * from (
        SELECT
        *,
        RANK() OVER (
            PARTITION BY username
            ORDER BY startdate DESC
        ) AS rk,
        COUNT(username) OVER (PARTITION BY username) AS cnt
        FROM UserActivity
        ) a
        WHERE a.rk = 2 OR a.cnt = 1;

""").display()
```
