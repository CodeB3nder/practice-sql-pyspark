```python
from pyspark.sql.functions import to_date
from pyspark.sql.types import StringType, StructType, StructField

# Failed table
failed_data = [
    ("2018-12-28",),
    ("2018-12-29",),
    ("2019-01-04",),
    ("2019-01-05",)
]
failed_schema = StructType([StructField("fail_date", StringType(), True)])
failed_df = spark.createDataFrame(failed_data, schema=failed_schema)
failed_df = failed_df.withColumn("fail_date", to_date("fail_date"))

# Succeeded table
succeeded_data = [
    ("2018-12-30",),
    ("2018-12-31",),
    ("2019-01-01",),
    ("2019-01-02",),
    ("2019-01-03",),
    ("2019-01-06",)
]
succeeded_schema = StructType([StructField("success_date", StringType(), True)])
succeeded_df = spark.createDataFrame(succeeded_data, schema=succeeded_schema)
succeeded_df = succeeded_df.withColumn("success_date", to_date("success_date"))

display(failed_df)
display(succeeded_df)
```

```python
succeeded_df.createOrReplaceTempView("succeeded")
failed_df.createOrReplaceTempView("failed")
# DBTITLE 1,Create a table
```
