```python
data = [
    ('India', 'SL', 'India'),
    ('SL', 'Aus', 'Aus'),
    ('SA', 'Eng', 'Eng'),
    ('Eng', 'NZ', 'NZ'),
    ('Aus', 'India', 'India')
]

columns = ['Team_1', 'Team_2', 'Winner']

icc_world_cup_df = spark.createDataFrame(data, columns)

display(icc_world_cup_df)
```

```python
icc_world_cup_df.createOrReplaceTempView('icc_world_cup')
```

```python
spark.sql("""
    select Team_1 as team, case when Team_1=Winner then 1 else 0 end as win_flag from icc_world_cup  
    union all 
    select Team_2 as team, case when Team_2=Winner then 1 else 0 end as win_flag from icc_world_cup  

""").display()
```

```python
from pyspark.sql.functions import *
```

```python
icc_world_cup_df.display()
```

```python
icc_world_cup_df.withColumnRenamed("Team_1", "team")
```

```python
team1_df = icc_world_cup_df.withColumnRenamed("Team_1", "team").withColumn('win_flag', when(col('team')==col('Winner'), 1).otherwise(0)).drop("Team_2")
team2_df = icc_world_cup_df.withColumnRenamed("Team_2", "team").withColumn('win_flag', when(col('team')==col('Winner'), 1).otherwise(0)).drop("Team_1")
team1_df.unionAll(team2_df).display()
```
