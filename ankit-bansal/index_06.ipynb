```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Initialize Spark Session
spark = SparkSession.builder.appName("SocialGraph").getOrCreate()

# Define schema for the person table
person_schema = StructType([
    StructField("PersonID", IntegerType(), True),
    StructField("Name", StringType(), True),
    StructField("Email", StringType(), True),
    StructField("Score", IntegerType(), True)
])

# Data from the image
person_data = [
    (1, "Alice", "alice2018@hotmail.com", 88),
    (2, "Bob", "bob2018@hotmail.com", 11),
    (3, "Davis", "davis2018@hotmail.com", 27),
    (4, "Tara", "tara2018@hotmail.com", 45),
    (5, "John", "john2018@hotmail.com", 63)
]

df_person = spark.createDataFrame(person_data, schema=person_schema)
df_person.show()

# Define schema for the friend table
friend_schema = StructType([
    StructField("PersonID", IntegerType(), True),
    StructField("FriendID", IntegerType(), True)
])

# Data from the image
friend_data = [
    (1, 2),
    (2, 1),
    (2, 3),
    (3, 5),
    (4, 2),
    (4, 3),
    (4, 5)
]

df_friend = spark.createDataFrame(friend_data, schema=friend_schema)
df_friend.show()
```

```python
df_person.createOrReplaceTempView("person")
df_friend.createOrReplaceTempView("friend")
```

```python
spark.sql("""
        with cte as (
    select f.PersonID, SUM(p.Score) as total_friend_score from
    friend f join person p on f.FriendID = p.PersonID 
    group by f.PersonID having total_friend_score > 100)

    select * from cte
""").display()
```
